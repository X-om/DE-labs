```python
import pandas as pd
from google.colab import files
uploaded = files.upload()

# creating a dictionory on data given by maa'm
data = {
    'Name': ['James', 'Barra', 'Sarah', 'Bill', 'Peter', 'Chloe', 'Ben', 'Anna'],
    'Gender': ['Male', 'Male', 'Female', 'Male', 'Male', 'Female', 'Male', 'Female'],
    'Age': [27, 32, 34, 23, 27, 32, 34, 23],
    'Weight': [75.1, 98.3, 63.5, 87.2, 75.1, 98.3, 63.5, 87.2],
    'Height': ['Short', 'Short', 'Medium', 'Tall', 'Tall', 'Short', 'Medium', 'Tall']
}
```

```python
#creating a DataFrame using pandas.DataFrame() method by provideing crated dict
df = pd.DataFrame(data)
print(df)
```

```python
# display all columns
df.columns
```

```python
# display the information about DataFrame such as Column , Non-null count , Datatypes
df.info()
```

```python
#creating a new column and asssigning all values of that column to null
df['country'] = None
print(df)
```

```python
# fillna method fills the null values with provided value , we can set the limits on changing values
df.fillna("india",inplace=True,limit = 5)
```

```python
print(df)
```

```python
# loc method insert the row at given location , follows 0 base indexing
df.loc[len(df)] = ["Om" , "Male", 20 , None , "Tall" ,  None]
df.loc[len(df)] = ["Abhi" , "Male", 21 , None , "short" ,  None]
```

```python
print(df)
```

```python
# mean() returns the statistical mean of a given column
mean_age = df['Age'].mean()
print(mean_age)
```

```python
# mode() returns the statistical mode of a given column
mode_age = df['Age'].mode()
print(mode_age)
```

```python
# median() returns the statistical median of a given column
median_age = df['Age'].median()
print(median_age)
```
