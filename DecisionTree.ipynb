#USER DEFINED FUNCTION

```python
import pandas as pd
import numpy as np
import math
from collections import Counter
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# Define the Decision Tree class
class DecisionTree:
    def __init__(self, max_depth=None):
        self.max_depth = max_depth
        self.tree = None

    def entropy(self, y):
        counts = np.bincount(y)
        probabilities = counts / len(y)
        entropy = -np.sum([p * math.log2(p) for p in probabilities if p > 0])
        return entropy

    def information_gain(self, y, left_y, right_y):
        parent_entropy = self.entropy(y)
        n = len(y)
        n_left, n_right = len(left_y), len(right_y)
        if n_left == 0 or n_right == 0:
            return 0
        child_entropy = (n_left / n) * self.entropy(left_y) + (n_right / n) * self.entropy(right_y)
        return parent_entropy - child_entropy

    def split(self, X, y, feature_index, threshold):
        left_indices = np.where(X[:, feature_index] <= threshold)
        right_indices = np.where(X[:, feature_index] > threshold)
        return X[left_indices], X[right_indices], y[left_indices], y[right_indices]

    def best_split(self, X, y):
        best_feature = None
        best_threshold = None
        best_gain = -1
        n_features = X.shape[1]

        for feature_index in range(n_features):
            thresholds = np.unique(X[:, feature_index])
            for threshold in thresholds:
                X_left, X_right, y_left, y_right = self.split(X, y, feature_index, threshold)
                if len(y_left) == 0 or len(y_right) == 0:
                    continue
                gain = self.information_gain(y, y_left, y_right)
                if gain > best_gain:
                    best_gain = gain
                    best_feature = feature_index
                    best_threshold = threshold

        return best_feature, best_threshold

    def build_tree(self, X, y, depth=0):
        n_samples, n_features = X.shape
        n_labels = len(np.unique(y))

        if n_labels == 1 or n_samples == 0 or (self.max_depth is not None and depth >= self.max_depth):
            leaf_value = self.most_common_label(y)
            return leaf_value

        feature_index, threshold = self.best_split(X, y)
        if feature_index is None:
            return self.most_common_label(y)

        X_left, X_right, y_left, y_right = self.split(X, y, feature_index, threshold)
        left_subtree = self.build_tree(X_left, y_left, depth + 1)
        right_subtree = self.build_tree(X_right, y_right, depth + 1)

        return {"feature_index": feature_index, "threshold": threshold, "left": left_subtree, "right": right_subtree}

    def predict_sample(self, x, tree):
        if not isinstance(tree, dict):
            return tree
        feature_index = tree["feature_index"]
        threshold = tree["threshold"]
        if x[feature_index] <= threshold:
            return self.predict_sample(x, tree["left"])
        else:
            return self.predict_sample(x, tree["right"])

    def predict(self, X):
        return [self.predict_sample(x, self.tree) for x in X]

    def fit(self, X, y):
        self.tree = self.build_tree(X, y)

    def most_common_label(self, y):
        counter = Counter(y)
        most_common = counter.most_common(1)[0][0]
        return most_common

# Load and prepare data
from google.colab import files
uploaded = files.upload()
df_net = pd.read_csv('Social_Network_Ads.csv')
X = df_net[['Age', 'EstimatedSalary']].values
y = df_net['Purchased'].values

# Visualize the distribution of EstimatedSalary
sns.histplot(df_net['EstimatedSalary'], kde=True)
plt.title('Salary Distribution')
plt.show()

# Heatmap for correlation between features
sns.heatmap(df_net.corr(), annot=True, cmap='coolwarm', square=True)
plt.title("Feature Correlation Heatmap")
plt.show()

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# Instantiate and fit the DecisionTree classifier
classifier = DecisionTree(max_depth=5)
classifier.fit(X_train, y_train)

# Predict on the test set
y_pred = classifier.predict(X_test)

# Display Predicted vs Actual values
results = np.concatenate((np.array(y_pred).reshape(len(y_pred), 1), y_test.reshape(len(y_test), 1)), axis=1)
print("Predicted vs Actual values:\n", results)

# Calculate and display accuracy and classification report
accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy: {accuracy * 100:.2f}%")
print(f"\nClassification Report:\n{classification_report(y_test, y_pred)}")
```
