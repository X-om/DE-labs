#**Tuple duplication on Dataframe**

```python
import numpy as np
import pandas as pd
import seaborn
import matplotlib
```

```python
student_data = {
    'id': [101, 102, 103, 104, 105, 106, 107, 108, 103, 102],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Hannah', 'Charlie', 'Bob'],
    'city': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Chicago', 'Los Angeles'],
    'branch': ['Science', 'Commerce', 'Science', 'Arts', 'Science', 'Commerce', 'Science', 'Arts', 'Science', 'Commerce'],
    'Maths': [85, 78, 92, 67, 88, 73, 95, 80, 92, 78],
    'Physics': [91, 82, 89, 72, 85, 77, 90, 83, 89, 82],
    'Chemistry': [87, 79, 93, 70, 86, 75, 94, 81, 93, 79]
}

df = pd.DataFrame(student_data)
```

```python
df
```

3) Creating another column in the dataframe which consists of the whole data in the row

```python
df['tuple'] = list(zip(df['id'], df['name'], df['city'], df['branch'], df['Maths'], df['Physics'], df['Chemistry']))
```

```python
df
```

4) Identify duplicates based on the 'tuple' column

```python
is_duplicate = df.duplicated(subset='tuple')
```

5)Dropping the duplicate rows using the '~' tilde sign.

```python
df_no_duplicates = df[~is_duplicate]
```

6) Dropping the extra column

```python
df_no_duplicates = df_no_duplicates.drop(columns=['tuple']).reset_index(drop=True)
```

```python
df_no_duplicates
```
